## Exploring population data
CSV file with World Bank data on population from http://api.worldbank.org/v2/en/indicator/SP.POP.TOTL?downloadformat=csv saved as ../data/population.csv

```python
# reading data using read_csv method
df_population = pd.read_csv('data/population.csv')
df_population.head()
```

### The read_csv() method fails, let's see what's wrong with the file

```python
# open the file and read first lines
file = open('data/population.csv')

#how many lines to read
numb_lines = 5

for i in range(numb_lines):
    line = file.readline()
    print(line)
    
file.close()
```

### As we see, first 4 lines are not the actual data in structured format

```python
# reading data using read_csv method, now skipping first 4 rows 
df_population = pd.read_csv('data/population.csv', skiprows=4, dtype='str')
df_population.head()
```

```python
# checking number of nulls in each column
print(df_population.isnull().sum())
```

```python
# checking number of nulls in each row
print(df_population.isnull().sum(axis=1))
```

### Let's drop the 'Unnamed: 64' column as it has only null values 

```python
# check to see if there are not null values
print(df_population['Unnamed: 64'].count())
```

```python
df_population = df_population.drop(columns=['Unnamed: 64'])
```

```python
# check how many rows have null values now
print(df_population[df_population.isnull().any(axis=1)].shape[0])
```
